```python
import geopandas as gpd
import pandas as pd
#pd.set_option('display.max_rows', None)
```

```python
ranking_file = 'ensemble_points_random_sample_20km_rankings_s2.fgb'
ranking = gpd.read_file(ranking_file)
len(ranking)
```

```python
ranking['rank_INATIDES'].describe()
```

```python
ranking.valid_perc.plot.hist();
```

```python
ranking[ranking.valid_perc>0.05].plot.scatter(x="worst_correlation", y="top_correlation")
```

```python
ranking[(ranking.top_correlation-ranking.worst_correlation)>0.1]
```

```python
# count how many times a model is ranked top
ranking[['top_model','top_correlation']].groupby('top_model').count()
```

```python
# mean correlation for when a model is ranked top
ranking[['top_model','top_correlation']].groupby('top_model').mean()
```

```python
# count how many times a model is ranked lowest
ranking[['worst_model','worst_correlation']].groupby('worst_model').count()
```
